```python
from matplotlib import pyplot as plt
import numpy as np

with plt.xkcd():
    y = [4, 14, 98, 657, 3624]
    ax1 = plt.axes(yscale='log')
    ax1.plot(range(len(y)), y)
    yticks = [10, 20, 50, 100, 200, 500, 1000, 2000, 5000]
    ax1.set_yticks(yticks)
    ax1.set_yticklabels(yticks)
    ax1.set_title("Round Number Log Scale")
    plt.savefig('/home/dmertz/Pictures/round-log.png')
```
